```python
height, width = 96, 96
```

```python
training = pd.read_csv('../data/training.csv')
```

```python
training_sample = training.loc[0]
training_sample_image_str = training_sample['Image']
training_sample_image = np.fromstring(training_sample_image_str, dtype=np.uint8, sep=' ')
training_sample_image = training_sample_image.reshape((height, width))
# cv2.line
left_eye_center_x = training_sample['left_eye_center_x']
left_eye_center_y = training_sample['left_eye_center_y']
right_eye_center_x = training_sample['right_eye_center_x']
right_eye_center_y = training_sample['right_eye_center_y']
left_eye_inner_corner_x = training_sample['left_eye_inner_corner_x']
left_eye_inner_corner_y = training_sample['left_eye_inner_corner_y']
left_eye_outer_corner_x = training_sample['left_eye_outer_corner_x']
left_eye_outer_corner_y = training_sample['left_eye_outer_corner_y']
right_eye_inner_corner_x = training_sample['right_eye_inner_corner_x']
right_eye_inner_corner_y = training_sample['right_eye_inner_corner_y']
right_eye_outer_corner_x = training_sample['right_eye_outer_corner_x']
right_eye_outer_corner_y = training_sample['right_eye_outer_corner_y']
left_eyebrow_inner_end_x = training_sample['left_eyebrow_inner_end_x']
left_eyebrow_inner_end_y = training_sample['left_eyebrow_inner_end_y']
left_eyebrow_outer_end_x = training_sample['left_eyebrow_outer_end_x']
left_eyebrow_outer_end_y = training_sample['left_eyebrow_outer_end_y']
right_eyebrow_inner_end_x = training_sample['right_eyebrow_inner_end_x']
right_eyebrow_inner_end_y = training_sample['right_eyebrow_inner_end_y']
right_eyebrow_outer_end_x = training_sample['right_eyebrow_outer_end_x']
right_eyebrow_outer_end_y = training_sample['right_eyebrow_outer_end_y']
nose_tip_x = training_sample['nose_tip_x']
nose_tip_y = training_sample['nose_tip_y']
mouth_left_corner_x = training_sample['mouth_left_corner_x']
mouth_left_corner_y = training_sample['mouth_left_corner_y']
mouth_right_corner_x = training_sample['mouth_right_corner_x']
mouth_right_corner_y = training_sample['mouth_right_corner_y']
mouth_center_top_lip_x = training_sample['mouth_center_top_lip_x']
mouth_center_top_lip_y = training_sample['mouth_center_top_lip_y']
mouth_center_bottom_lip_x = training_sample['mouth_center_bottom_lip_x']
mouth_center_bottom_lip_y = training_sample['mouth_center_bottom_lip_y']

training_sample_image = cv2.circle(training_sample_image, center=(int(left_eye_center_x), int(left_eye_center_y)), radius=1, color=255, thickness=1)
training_sample_image = cv2.circle(training_sample_image, center=(int(right_eye_center_x), int(right_eye_center_y)), radius=1, color=255, thickness=1)
training_sample_image = cv2.circle(training_sample_image, center=(int(left_eye_inner_corner_x), int(left_eye_inner_corner_y)), radius=1, color=255, thickness=1)
training_sample_image = cv2.circle(training_sample_image, center=(int(left_eye_outer_corner_x), int(left_eye_outer_corner_y)), radius=1, color=255, thickness=1)
training_sample_image = cv2.circle(training_sample_image, center=(int(right_eye_inner_corner_x), int(right_eye_inner_corner_y)), radius=1, color=255, thickness=1)
training_sample_image = cv2.circle(training_sample_image, center=(int(right_eye_outer_corner_x), int(right_eye_outer_corner_y)), radius=1, color=255, thickness=1)
training_sample_image = cv2.circle(training_sample_image, center=(int(left_eyebrow_inner_end_x), int(left_eyebrow_inner_end_y)), radius=1, color=255, thickness=1)
training_sample_image = cv2.circle(training_sample_image, center=(int(left_eyebrow_outer_end_x), int(left_eyebrow_outer_end_y)), radius=1, color=255, thickness=1)
training_sample_image = cv2.circle(training_sample_image, center=(int(right_eyebrow_inner_end_x), int(right_eyebrow_inner_end_y)), radius=1, color=255, thickness=1)
training_sample_image = cv2.circle(training_sample_image, center=(int(right_eyebrow_outer_end_x), int(right_eyebrow_outer_end_y)), radius=1, color=255, thickness=1)
training_sample_image = cv2.circle(training_sample_image, center=(int(nose_tip_x), int(nose_tip_y)), radius=1, color=255, thickness=1)
training_sample_image = cv2.circle(training_sample_image, center=(int(mouth_left_corner_x), int(mouth_left_corner_y)), radius=1, color=255, thickness=1)
training_sample_image = cv2.circle(training_sample_image, center=(int(mouth_right_corner_x), int(mouth_right_corner_y)), radius=1, color=255, thickness=1)
training_sample_image = cv2.circle(training_sample_image, center=(int(mouth_center_top_lip_x), int(mouth_center_top_lip_y)), radius=1, color=255, thickness=1)
training_sample_image = cv2.circle(training_sample_image, center=(int(mouth_center_bottom_lip_x), int(mouth_center_bottom_lip_y)), radius=1, color=255, thickness=1)


plt.imshow(training_sample_image, cmap='gray')
plt.axis('off')
# print('training_sample_image.shape:', training_sample_image.shape)
```
